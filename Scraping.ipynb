```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from IPython.display import display

# Ganti dengan Client ID dan Client Secret Anda
client_id = "Masukkan client id"
client_secret = "Masukkan client secret"
```

```python
# Input manual nama artis yang ingin di-scrape
artist_name = input("Masukkan nama artis yang ingin di-scrape: ")
```

```python
# Inisialisasi Spotipy dengan kredensial
credentials = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=credentials)

# Fungsi untuk scraping album dan judul lagu dari artis
def scrape_albums_and_tracks(artist_id):
    albums = sp.artist_albums(artist_id, album_type='album')
    album_info = []

    for album in albums['items']:
        # Ambil informasi album
        album_details = {
            'album_name': album['name'],
            'release_date': album['release_date'],
            'album_url': album['external_urls']['spotify'],
            'tracks': []
        }

        # Ambil detail lagu dari album
        tracks = sp.album_tracks(album['id'])
        for track in tracks['items']:
            # Ambil durasi lagu dan artis
            track_details = {
                'track_name': track['name'],
                'track_url': track['external_urls']['spotify'],
                'duration_ms': track['duration_ms'],
                'artists': [artist['name'] for artist in track['artists']]  # Ambil nama artis
            }

            # Mendapatkan popularitas lagu dengan ID track
            track_info = sp.track(track['id'])
            track_details['popularity'] = track_info['popularity']  # Popularitas lagu

            album_details['tracks'].append(track_details)

        album_info.append(album_details)
    return album_info

# Mencari artis berdasarkan nama
results = sp.search(q=artist_name, type='artist')
if results['artists']['items']:
    artist = results['artists']['items'][0]  # Ambil artis pertama yang ditemukan
    print(f"Artis ditemukan: {artist['name']} (ID: {artist['id']})")

    # Melakukan scraping album dan lagu dari artis
    albums = scrape_albums_and_tracks(artist['id'])

    # Menyiapkan data untuk DataFrame
    data = []

    for album in albums:
        for track in album['tracks']:
            # Konversi durasi dari milidetik ke menit:detik
            duration = track['duration_ms'] / 1000
            minutes, seconds = divmod(duration, 60)
            duration_str = f"{int(minutes)}:{int(seconds):02d}"  # Format 00:00

            # Menambahkan data ke list
            data.append({
                'Album Name': album['album_name'],
                'Release Date': album['release_date'],
                'Track Name': track['track_name'],
                'Track URL': track['track_url'],
                'Duration': duration_str,
                'Artists': ', '.join(track['artists']),
                'Popularity': track['popularity'],
                'Album URL': album['album_url']
            })

    # Membuat DataFrame dari data
    df = pd.DataFrame(data)

    # Menampilkan DataFrame
    print("\nData Hasil Scraping:")
    display(df)  # Menggunakan display untuk format yang lebih baik di Jupyter Notebook
else:
    print("Artis tidak ditemukan.")
```

```python
# Ekspor DataFrame ke CSV
df.to_csv('hasil_scraping.csv', index=False)

# Import modul files untuk mendownload file
from google.colab import files

# Download file CSV
files.download('hasil_scraping.csv')
```
